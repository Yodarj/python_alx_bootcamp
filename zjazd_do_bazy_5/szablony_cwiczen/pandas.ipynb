```python
df.query('''home_team == 'Newcastle United' or away_team == 'Newcastle United' ''')
```

```python
team = 'Newcastle United'
```

```python
team_matches = df[(df.home_team == team) | (df.away_team == team)]
```

```python
team_matches.reset_index()
```

```python
def team_goals(x, team):
    if x.home_team == team:
        return x.home_team_goals
    elif x.away_team == team:
        return x.home_team_goals
```

```python
team_matches['team_goals'] = team_matches.apply(lambda x: x.home_team_goal if x.home_team == team else x.away_team_goal, axis=1)
```

```python
team_matches
```

```python
team_matches['opponents_goals'] = team_matches.apply(lambda x: x.away_team_goal if x.home_team == team else x.home_team_goal, axis=1)
```

```python
team_matches
```

```python
team_matches.sum()
```

```python
goal_per_season = team_matches.groupby('season').agg({'team_goals': sum, 'opponents_goals': sum})
```

```python
goal_per_season
```

```python
@matplotlib inline
```

```python
goal_per_season.plot(kind="bar")
```
